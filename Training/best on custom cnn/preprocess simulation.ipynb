```python
import os
import cv2
import matplotlib.pyplot as plt

# Image size
IMG_SIZE = (128, 128)

def preprocess_and_visualize(image_path):
    try:
        # Load the original image
        original_image = cv2.imread(image_path)
        if original_image is None:
            raise FileNotFoundError(f"Image file not found or could not be read: {image_path}")

        # Create a figure with subplots arranged vertically
        plt.figure(figsize=(6, 12))

        # Display original image
        plt.subplot(5, 1, 1)
        plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
        plt.title('Original Image')
        plt.axis('off')

        # Convert to grayscale
        gray = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
        plt.subplot(5, 1, 2)
        plt.imshow(gray, cmap='gray')
        plt.title('Grayscale')
        plt.axis('off')

        # Histogram equalization
        equalized = cv2.equalizeHist(gray)
        plt.subplot(5, 1, 3)
        plt.imshow(equalized, cmap='gray')
        plt.title('Equalized Histogram')
        plt.axis('off')

        # Gaussian blur
        gaussian_filtered = cv2.GaussianBlur(equalized, (5, 5), 0)
        plt.subplot(5, 1, 4)
        plt.imshow(gaussian_filtered, cmap='gray')
        plt.title('Gaussian Blurred')
        plt.axis('off')

        # Resize image
        resized_image = cv2.resize(gaussian_filtered, IMG_SIZE)
        plt.subplot(5, 1, 5)
        plt.imshow(resized_image, cmap='gray')
        plt.title('Resized to 299x299')
        plt.axis('off')

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Error processing {image_path}: {str(e)}")

# Example usage:
image_path = r"H:\Final Dataset\cow_235\cow_235_1.jpg"
preprocess_and_visualize(image_path)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.optimizers import Adam

# Define your constants
IMG_SIZE = (128, 128)  # Image size as per your preprocessing
num_classes = 105  # Number of output classes

# Define the model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_SIZE[0], IMG_SIZE[1], 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Print the model summary
model.summary()

# Optional: Generate the model architecture diagram
from keras.utils.vis_utils import plot_model

plot_model(model, to_file='model_architecture.png', show_shapes=True)

# Now you can proceed to fit this model to your data
# Example:
# model.fit(X_train, y_train, epochs=epochs, validation_data=(X_val, y_val))
```
